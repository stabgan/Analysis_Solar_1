# The generation of DC current (A) through the sensor is directly proportional to the irridation of the solar sensor (W/m^2) , that's why the graph is similar .

```python
amp1 = td['WRHM6Y94'].tolist()
amp2 = td['WRHM6Y94.3'].tolist()
amp3 = td['WRHM6Y94.4'].tolist()
amp4 = td['WRHM6Y94.7'].tolist()

y1 = list(range(len(amp1)-3))
y2 = list(range(len(amp2)-3))
y3 = list(range(len(amp3)-3))
y4 = list(range(len(amp4)-3))

plt.figure(num=None, figsize=(15, 4), dpi=80, facecolor='w', edgecolor='k')

plt.subplots_adjust(left=None, bottom=10, right=None, top=15, wspace=None, hspace=None)
plt.subplot(4, 1, 1)
plt.plot(y1 , amp1[3:])
plt.title('CURRENT THROUG 4 SENSORS')
plt.xlabel('Every 15 mins')
plt.ylabel('A.Ms.Amp')

plt.subplot(4, 1, 2)
plt.plot(y2 , amp2[3:])
plt.xlabel('Every 15 mins')
plt.ylabel('A1.Ms.Amp')

plt.subplot(4, 1, 3)
plt.plot(y3 , amp3[3:], 'b--')
plt.xlabel('Every 15 mins')
plt.ylabel('B.Ms.Amp')

plt.subplot(4, 1, 4)
plt.plot(y4 , amp4[3:], 'b--')
plt.xlabel('Every 15 mins')
plt.ylabel('B1.Ms.Amp')


plt.show()
```

# In the generation of the energy we can see the sudden rise in the generation of power and the sudden decrease when the sun starts to set 

```python
etot = td['WRHM6Y94.9'].tolist()
y = list(range(len(etot)-3))


plt.figure(num=None, figsize=(15, 4), dpi=80, facecolor='w', edgecolor='k')

plt.plot(y , etot[3:])
plt.title("TOTAL ENERGY GENERATED")
plt.xlabel("Watts generated")
plt.ylabel("Every 15 mins")
plt.show()
```

```python
phv1 = td['WRHM6Y94.15'][3:]
phv2 = td['WRHM6Y94.16'][3:]
phv3 = td['WRHM6Y94.17'][3:]
```

```python
def diff (x,y,z) :
    x,y,z = float(x) , float(y) , float(z)
    return max(x-y,y-z,x-z)

n = len(phv1)
l = []
for i in range(n) :
    l.append(diff(phv1[i],phv2[i],phv3[i]))
```

# The voltages should be equal for 3 phase lines in Ideal case . The lower the difference the better .

```python
plt.figure(num=None, figsize=(15, 4), dpi=80, facecolor='w', edgecolor='k')

plt.plot(list(range(n)),l)
plt.title("DIFFERENCE OF THE VOLTAGE GENERATED IN THREE PHASES")
plt.ylabel("difference")
plt.xlabel("Every 15 mins")
plt.show()
```

# The frequency of the 3 phase AC should be constant in the ideal cases . But there is fluctations in real case . The lower the fluctuations the better .

```python
fre = td['WRHM6Y94.14'].tolist()[3:]

plt.figure(num=None, figsize=(15, 4), dpi=80, facecolor='w', edgecolor='k')

plt.plot(list(range(len(fre))),fre)
plt.title("FREQUENCY OF THE 3 PHASE AC SUPPLY")
plt.ylabel("Hz")
plt.xlabel("Every 15 mins")
plt.show()
```
