```python
phv1 = pd.to_numeric(td['WRHM6Y94.15'], errors='coerce').values[3:]
phv2 = pd.to_numeric(td['WRHM6Y94.16'], errors='coerce').values[3:]
phv3 = pd.to_numeric(td['WRHM6Y94.17'], errors='coerce').values[3:]
```

```python
def voltage_diff(x, y, z):
    """Return the maximum pairwise difference among three phase voltages."""
    x, y, z = float(x), float(y), float(z)
    return max(abs(x - y), abs(y - z), abs(x - z))

n = len(phv1)
phase_diffs = [voltage_diff(phv1[i], phv2[i], phv3[i]) for i in range(n)]
```

# The voltages should be equal for 3-phase lines in the ideal case. The lower the difference, the better.

```python
plt.figure(num=None, figsize=(15, 4), dpi=80, facecolor='w', edgecolor='k')
plt.plot(list(range(n)), phase_diffs)
plt.title("DIFFERENCE OF THE VOLTAGE GENERATED IN THREE PHASES")
plt.ylabel("Voltage difference (V)")
plt.xlabel("Every 15 mins")
plt.tight_layout()
plt.show()
```

# The frequency of the 3-phase AC should be constant in the ideal case, but there are fluctuations in practice. The lower the fluctuations, the better.

```python
fre = pd.to_numeric(td['WRHM6Y94.14'], errors='coerce').values[3:]

plt.figure(num=None, figsize=(15, 4), dpi=80, facecolor='w', edgecolor='k')
plt.plot(list(range(len(fre))), fre)
plt.title("FREQUENCY OF THE 3-PHASE AC SUPPLY")
plt.ylabel("Hz")
plt.xlabel("Every 15 mins")
plt.tight_layout()
plt.show()
```
